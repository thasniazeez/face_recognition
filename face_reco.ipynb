```python
import cv2
```

```python
def name_map(x):
    if x=="sachin":
        return 0
    if x=="obama":
        return 1
    if x=="messi":
        return 2
```

```python
def get_name(x):
    if x==0:
        return "sachin"
    if x==1:
        return 'obama'
    if x==2:
        return "messi"
```

```python
def detect_face(img):
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face_cascade=cv2.CascadeClassifier(r"D:\DL Projects\face_recognition\haarcascade_frontalface_default.xml")
    face=face_cascade.detectMultiScale(gray,minNeighbors=10)
    if len(face)==0:
        return None,None
    x,y,w,h=face[0]   #frst elemt lle value x,y,w,h llot pokm
    return gray[x:x+w,y:y+h],face[0]
```

```python
import os
```

```python
def prepare_training_data(data_folder):
    dirs=os.listdir(data_folder)
    print(dirs)
    faces=[]
    labels=[]
    for dir_name in dirs :  #dir=messi
        sub_dir=data_folder+'/'+dir_name
        # print(sub_dir)
        images=os.listdir(sub_dir)
        print(images)
        for img in images:
            imgpath=sub_dir+'/'+img
            # print(imgpath)
            image=cv2.imread(imgpath)
            cv2.imshow("training image",image)
            cv2.waitKey(500)
            # detect
            face,rect=detect_face(image)
            if face is not None:
                faces.append(face)
                labels.append(name_map(dir_name))
                print(labels)
                
            
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    return faces,labels
```

```python

prepare_training_data(r"D:\DL Projects\face_recognition\DB")


```

```python
faces,labels=prepare_training_data(r"D:\DL Projects\face_recognition\DB")
print(labels)
```

```python
len(labels)
```

```python
recognizer=cv2.face.LBPHFaceRecognizer_create()
recognizer
```

```python
# converting in t numpy
import numpy as np
```

```python
recognizer.train(faces,np.array(labels))
```

```python
# predictions
def draw_rectangle(img,rect):
    (x,y,w,h)=rect
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),2)
    
```

```python
def write_text(img,text,x,y):
    cv2.putText(img,text,(x,y),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
    # a person face and the name of the persons face
```

```python
def predict_new(test_img):
    img=test_img
    face,rect=detect_face(test_img)
    label=recognizer.predict(face)
    # print(label)
    label_text=get_name(label[0])
    print(label_text)
    draw_rectangle(img,rect)
    write_text(img,label_text,rect[0],rect[1])
    return (img)
```

```python
test_img=cv2.imread(r"D:\DL Projects\face_recognition\WhatsApp Image 2024-01-02 at 14.17.20_f50790d4.jpg")
predicted=predict_new(test_img)
cv2.imshow("IMAGE",predicted)
cv2.waitKey()
cv2.destroyAllWindows()
```
